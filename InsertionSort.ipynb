```python
def insertion_sort(array):
    if(len(array) == 0):
        return array
    
    for i in range(len(array)):
        j = i
        while(j > 0 and array[j] < array[j - 1]):
            temp = array[j - 1]
            array[j - 1] = array[j]
            array[j] = temp
            j -= 1
    return array
```

```python
def merge_sort(array):
    if(len(array) == 1):
        return array
    else:
        split = len(array) // 2
        array1 = merge_sort(array[:split])
        array2 = merge_sort(array[split:])
        return merge(array1, array2)

def merge(array1, array2):
    i, j = 0, 0
    merged_array = []
    while(i < len(array1) and j < len(array2)):
        if(array1[i] <= array2[j]):
            merged_array.append(array1[i])
            i += 1
        else:
            merged_array.append(array2[j])
            j += 1
    while(i < len(array1)):
        merged_array.append(array1[i])
        i += 1
    while(j < len(array2)):
        merged_array.append(array2[j])
        j += 1
    return merged_array
```

```python
import random

def randomArrayGenerator(n):
    array = []
    for i in range(n):
        array.append(random.randint(0, 1000000))
    return array
        
```

```python
import datetime

def testSortAlgorithm(n):
    timesOfInsert = []
    timesOfMerge = []
    for i in range(1, n+1):
        randomArray = randomArrayGenerator(i)
        startingtime = datetime.datetime.now()
        insertion_sort(randomArray)
        endtime = datetime.datetime.now()
        timesOfInsert.append((endtime - startingtime).microseconds)
        startingtime = datetime.datetime.now()
        merge_sort(randomArray)
        endtime = datetime.datetime.now()
        timesOfMerge.append((endtime - startingtime).microseconds)
    return timesOfInsert, timesOfMerge
```

```python
testSortAlgorithm(500)
```

```python
def quicksort(array, start, end):
    if start >= end:
        return
    else:
#         pivotIndex = partition(array, start, end)
#         array[start], array[pivotIndex] = array[pivotIndex], array[start]
        newPivotIndex = partition(array, start, end)
        quicksort(array, start, newPivotIndex - 1)
        quicksort(array, newPivotIndex + 1, end)

def partition(array, start, end):
    pivot = array[start]
    i = start + 1
    for j in range(start + 1, end + 1):
        if(array[j] < pivot):
            array[i], array[j] = array[j], array[i]
            i += 1
    array[start], array[i - 1] = array[i - 1], array[start]
    return i - 1


def quickSortAlgorithm(array):
    quicksort(array, 0, len(array) - 1)
```

```python
import datetime

def compareQuickAndMergeSort(n):
    timesOfQuick = []
    timesOfMerge = []
    for i in range(1, n+1):
        randomArray = randomArrayGenerator(i)
        startingtime = datetime.datetime.now()
        quickSortAlgorithm(randomArray)
        endtime = datetime.datetime.now()
        timesOfQuick.append((endtime - startingtime).microseconds)
        startingtime = datetime.datetime.now()
        merge_sort(randomArray)
        endtime = datetime.datetime.now()
        timesOfMerge.append((endtime - startingtime).microseconds)
    return timesOfQuick, timesOfMerge
```
